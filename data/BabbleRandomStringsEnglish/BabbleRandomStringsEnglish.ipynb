```python
import random
import pandas as pd
import numpy as np
```

# get the list of segments

```python
segs = pd.read_csv("./../MakingOverTrainSet/WordsToBeScored_11.5.2023.txt", header = None)[0]
phonemes = ['B',
'CH',
'D',
'DH',
'F',
'G',
'JH',
'K',
'P',
'S',
'SH',
'T',
'TH',
'V',
'Z',
'ZH',
'HH',
'M',
'N',
'NG',
'L',
'R',
'W',
'Y',
'IY',
'IH',
'EH',
'EY',
'AE',
'AA',
'AW',
'AY',
'AO',
'OY',
'OW',
'UH',
'UW',
'ER',
'AH'
]

vowels = ['IY',
'IH',
'EH',
'EY',
'AE',
'AA',
'AW',
'AY',
'AO',
'OY',
'OW',
'UH',
'UW',
'ER',
'AH']

consonants = ['B',
'CH',
'D',
'DH',
'F',
'G',
'JH',
'K',
'P',
'S',
'SH',
'T',
'TH',
'V',
'Z',
'ZH',
'HH',
'M',
'N',
'NG',
'L',
'R',
'W',
'Y']

syllable_types = ['V',
'CV',
'CCV',
'CCCV',
'CVC',
'CCVC',
'CCCVC',
'CVCC',
'CCVCC',
'CCCVCC',
'CVCCC',
'CCVCCC',
'CCCVCCC',
'CVCCCC',
'CCVCCCC',
'CCCVCCCC',
]
```

```python
def babble_random_segments(num_words,phonemes):
    gen_words = []
    for _ in range(num_words):
        length_in_segs = np.random.poisson(5)
        if length_in_segs == 0:
            length_in_segs = 1
        else:
            pass
        #print(length_in_segs)
        w = " ".join(random.choices(phonemes,k=length_in_segs))
        #print(w)
        gen_words.append(w)
    return gen_words

def babble_random_well_formed_syllables(num_words, syllable_types, consonants, vowels):
    gen_words = []
    for _ in range(num_words):
        gen_word_so_far = ""
        num_sylls = np.random.poisson(2)
        if num_sylls == 0:
            num_sylls = 1
        else:
            pass
        for syll in range(num_sylls):
            current_syllable = random.choices(syllable_types, k = 1)
            for struc in current_syllable[0]:
                if struc == "C":
                    gen_word_so_far += random.choices(consonants,k=1)[0]
                    gen_word_so_far += " "
                elif struc == "V":
                    gen_word_so_far += random.choices(vowels,k=1)[0]
                    gen_word_so_far += " "
                else:
                    raise NotImplementedError("syllable shapes must be all and only Cs or Vs.")
        gen_words.append(gen_word_so_far.strip())
    return gen_words

```

```python
# babble random segmetns

            
result = babble_random_segments(10000,phonemes)
print(result)
outfile = open("babble_10k_random_strings.txt","w")
for item in result:
    #print(item)
    outfile.write(item+"\n")
outfile.close()
```

```python
# babble random well-formed syllable types

            
result = babble_random_well_formed_syllables(10000,syllable_types,consonants,vowels)
print(result)
outfile = open("babble_10k_random_well_formed_syllables.txt","w")
for item in result:
    #print(item)
    outfile.write(item+"\n")
outfile.close()
```
